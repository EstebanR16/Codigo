1) Abrir el archivo "Emisiones_CO2.csv" y cargar sus datos en un diccionario.

```python
import os
archivo = open('Emisiones_CO2.csv', 'r',encoding='1256')
```

```python
dicc_emisiones = {  'cod_pais' : [],
                    'nom_pais' : [],
                    'region' : [],
                    'anio' : [],
                    'co2' : [],
                    'co2_percapita' : []}
```

```python
next(archivo)
for linea in archivo:
  linea=linea.strip()
  camp=linea.split("|")
  dicc_emisiones['cod_pais'].append(camp[0])
  dicc_emisiones['nom_pais'].append(camp[1])
  dicc_emisiones['region'].append(camp[2])
  dicc_emisiones['anio'].append(camp[3])
  dicc_emisiones['co2'].append(camp[4])
  dicc_emisiones['co2_percapita'].append(camp[5])

  
```

```python
archivo.close()
```

```python
def normalizacion(lista):
  for indice, elem in enumerate(lista):
    if elem:
      elem= elem.replace('.','')
      elem= elem.replace(',','.')
      elem=float(elem)
    
    else: elem=None

    lista[indice]= elem
  return lista

```

```python
dicc_emisiones['co2']=normalizacion(dicc_emisiones['co2'])
dicc_emisiones['co2_percapita']=normalizacion(dicc_emisiones['co2_percapita'])
```

```python
for i in dicc_emisiones['co2']:
  print(i)
```

```python
for i in range(len(dicc_emisiones['anio'])):
  dicc_emisiones['anio'][i]= int(dicc_emisiones['anio'][i])
```

```python
filtro_region= 'América Latina y Caribe'
filtro_anio=2010
emisiones=0
for indice, elemento in enumerate(dicc_emisiones['region']):
  if dicc_emisiones['region'][indice]==filtro_region and dicc_emisiones['anio'][indice]==filtro_anio and dicc_emisiones['co2'][indice]!=None: 
    emisiones +=dicc_emisiones['co2'][indice]
print(f'Las emiciones en: {filtro_region} para el año {filtro_anio} fueron de {round(emisiones,3)} kt')

```
